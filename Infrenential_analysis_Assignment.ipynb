 **9)Convert the normal distribution to standard normal distribution for
salary**

```python
def std_deviation(column_name):
  import seaborn as sns
  mean=column_name.mean()
  std=column_name.std()
  change_in_list=[i for i in column_name]
  z_score=[((j-mean)/std) for j in change_in_list]
  sns.distplot(z_score)
  return sum(z_score)/len(z_score)
```

```python
std_deviation(placed_data['salary'])
```


**10)What is the probability Density Function of the salary range from
700000 to 900000?**

```python
import seaborn as sns
from matplotlib import pyplot
from scipy.stats import norm
```

```python
def get_pdf(dataset,startrange,endrange):
  ax=sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
  pyplot.axvline(startrange,color='Red')
  pyplot.axvline(endrange,color='Red')
  sample=dataset
  sample_mean=sample.mean()
  sample_std=sample.std()
  dist=norm(sample_mean,sample_std)
  values=[i for i in range(startrange,endrange)]
  probality=[dist.pdf(value) for value in values]
  prob=sum(probality)
  print('xx',prob)
  return prob
```

```python
get_pdf(placed_data['salary'],700000,900000)
```
