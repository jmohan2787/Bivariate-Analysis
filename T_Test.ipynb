Independent sample--> Unpaired

Different Group(male,Female) but same condition

```python
from scipy.stats import ttest_ind
male=data[data['gender']=='M']['salary']
female=data[data['gender']=='F']['salary']
ttest_ind(male,female)
```

Dependend sample --> Paired

Same Group(Male) but different Condition(ssc_p,hsc_p)

```python
from scipy.stats import ttest_rel
male=data[data['gender']=='M']['ssc_p']
male1=data[data['gender']=='M']['hsc_p']
ttest_rel(male,male1)
```
